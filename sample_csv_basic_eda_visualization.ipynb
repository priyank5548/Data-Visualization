converting salary column from object to float

```python
def clean_salary(salary):
  if isinstance(salary, str):
    return float(salary.replace('$','').replace(',','')) # "10000" -> 10000
  return float(salary)

df['Salary'] = df['Salary'].apply(clean_salary)
```

```python
df['Salary']
```

```python
df.groupby('gender')['Salary'].mean()
```

```python
def male_salary(salary):
  if salary['gender'] == 'Male':
    return salary['Salary']

def female_salary(salary):
  if salary['gender'] == 'Female':
    return salary['Salary']
```

```python
df['male_salary'] = df.apply(male_salary, axis=1)
df['female_salary'] = df.apply(female_salary,axis=1)
```

```python
df.head()
```

```python
df.groupby('Job Title')[['Salary', 'male_salary', 'female_salary']].mean().reset_index()
```

```python
df.groupby(['Job Title', 'City'])[['Salary', 'male_salary', 'female_salary']].mean().reset_index()
```

Handling Missing data

mean, median, mode

```python
df.head()
```

```python
# df['City'].fillna('London', inplace=True) # old method
df.fillna({'City':'London'}, inplace=True) # new method
```

```python
median = df['Latitude'].median()
df.fillna({'Latitude': median}, inplace=True)
```

removing rows conataining NaN value

```python
df.dropna(subset=['Job Title'],inplace=True)
```

Removing Duplicates

```python
df.drop_duplicates(inplace=True)
df.drop_duplicates(subset=['first_name', 'Job Title'], inplace=True)
```
