```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# Line Plot
plt.plot(data['total_bill'])
plt.title("Total Bill Trend")
plt.xlabel("Customer Index")
plt.ylabel("Bill Amount")
plt.show()
```

```python
#Scatter Chart
plt.scatter(data['total_bill'],data['tip'])
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
#Bar Chart

data['day'].value_counts().plot(kind='bar')
plt.title("Customers per day")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()
```

```python
from matplotlib.lines import lineStyles
plt.figure(figsize=(10,5))
plt.plot(data['total_bill'], color = 'green',marker='o',linestyle='--')
plt.title("Styled line plot")
plt.xlabel("Index")
plt.ylabel("Total Bill")
plt.grid(True)
```

```python
#Histrogram distribution
plt.hist(data['total_bill'],bins=20)
plt.title("Bill Distribution")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
```
